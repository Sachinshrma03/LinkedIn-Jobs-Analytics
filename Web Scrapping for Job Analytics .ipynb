```python
#Importing the required Libraries

import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# Activating the Chrome driver.

browser = webdriver.Chrome('chromedriver.exe')
browser.get("https://www.linkedin.com")
```

```python
# Inserting the Login Details into the page.

username = browser.find_element("id","session_key")
username.send_keys("sachinshrma03@gmail.com")
password = browser.find_element("id","session_password")
password.send_keys("Password")
```

```python
# Clicking on the login Button.

login_button = browser.find_element(By.CLASS_NAME, "sign-in-form__submit-button")
login_button.click()
```

```python
# Recording all the required links into a list.

linklist=[]
limit = 0
while limit<100:
    finance='https://www.linkedin.com/jobs/search/?alertAction=viewjobs&currentJobId=3247208485&geoId=102713980&keywords=finance&location=India&refresh=true&start='+str(limit)
    marketing='https://www.linkedin.com/jobs/search/?alertAction=viewjobs&currentJobId=3261611806&geoId=102713980&keywords=marketing&location=India&refresh=true&start='+str(limit)
    software='https://www.linkedin.com/jobs/search/?alertAction=viewjobs&currentJobId=3243881139&geoId=102713980&keywords=software&location=India&refresh=true&start='+str(limit)
    dataanalyst ='https://www.linkedin.com/jobs/search/?currentJobId=3247212696&geoId=105214831&keywords=analyst&location=Bengaluru%2C%20Karnataka%2C%20India&refresh=true&start'+str(limit)
    linklist.append(finance)
    linklist.append(marketing)
    linklist.append(software)
    linklist.append(dataanalyst)
    limit+=25
```

```python
# Checking the Number of Links into the linklist by checking the length of the list.

len(linklist)
```

```python
# Listing the Href links of all the links that we have stored into a list Link.

link = []
for i in linklist:
    browser.get(i)
    time.sleep(5)
    links = browser.find_elements('tag name','a')
    for i in links:
        link.append(i.get_attribute('href'))
```

```python
# Checking the number of href links that we stored into the list.

len(link)
```

```python
# Checking the links 

link
```

```python
# Only fetching the href links having Jobs with the filter of 'jobs/view' into the Href List.

href=[]
for i in link:
    try:
        if 'jobs/view' in i:
            href.append(i)
    except:
        pass
```

```python
# Checking the number of href links that are there into href list.

len(href)
```

```python
# Eliminating the repeated href links by creating a set of the href list and assigning to a new variable name.
# Checking the number of links in the Href_final by checking it's length.

href_final=set(href)
len(href_final)
```

```python
# Typecasting the href_final to a List.

href_final = list(href_final)
```

```python
# Verifying the number of links.

len(href_final)
```

```python
# Iterating the href_final links and storing in respective lists to get the required details like Job Title, Locations, Company name, Followers on linkedin,
# Employee count, Job opening level and number of applicants.

job = []
location= []
company=[]
followers=[]
emp_details=[]
inv_level=[]
applicants=[]

for i in range(0,10):
    browser.get(href_final[i])
    time.sleep(10)
    try:
        j = browser.find_element("class name","jobs-unified-top-card__job-title")
        job.append(j.text)
    except:
        job.append('NA')
    try:
        l = browser.find_element("class name","jobs-unified-top-card__bullet")
        location.append(l.text)
    except:
        location.append('NA')
    try:
        c = browser.find_element("class name","jobs-unified-top-card__company-name")
        company.append(c.text)
    except:
        company.append('NA')
    try:
        i = browser.find_elements(By.CLASS_NAME,"jobs-unified-top-card__job-insight")
        inv_level.append(i[0].text)
        emp_details.append(i[1].text)
    except:
        inv_level.append('NA')
        emp_details.append('NA')
    try:
        f = browser.find_element("class name","artdeco-entity-lockup__subtitle")
        followers.append(f.text)
    except:
        followers.append('NA')
    try:
        ap = browser.find_element("class name","jobs-unified-top-card__applicant-count")
        applicants.append(ap.text)
    except:
        applicants.append('NA')
```

```python
# Checking the details stored in the lists of required information.

print(len(job))
print(len(location))
print(len(company))
print(len(emp_details))
print(len(followers))
print(len(inv_level))
print(len(applicants))
```

```python
# Transforming the scraped data into a Pandas dataframe.

col=['Job Title','Company Name','Location','Followers', 'Job Details','Empolyees', 'Involvement','Applicants']
df=pd.DataFrame({'Job': job,'Company Name':company,'Location':location,'Followers':followers,'Employees':emp_details,'Involvement':inv_level,'Applicants':applicants})
df
```

```python
# Saving pandas DataFrame as a CSV.

df.to_csv('JobAnalytics.csv')
```
